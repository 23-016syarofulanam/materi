**penjelasan tentang mengukur Jarak dengan tipe data campuran**

Mengukur jarak untuk tipe data campuran dilakukan karena dalam satu database sering kali terdapat berbagai jenis atribut sekaligus, seperti Nominal, Symmetric Binary, Asymmetric Binary, Numeric, dan Ordinal

**penjelasan secara mekanisme**

.Untuk mendapatkan satu nilai jarak (dissimilarity) dari berbagai tipe data, kita menggunakan metode pembobotan. Rumus umumnya menggabungkan kontribusi dari setiap atribut f untuk pasangan objek i dan j

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from scipy.spatial.distance import pdist, squareform

# 1. Load Data Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Mengambil 5 sampel pertama untuk visualisasi
subset_data = df.head(5)

# 2. Perhitungan Jarak (Minkowski Distance)
# a. Manhattan Distance (h=1)
dist_manhattan = squareform(pdist(subset_data, metric='cityblock'))

# b. Euclidean Distance (h=2)
dist_euclidean = squareform(pdist(subset_data, metric='euclidean'))

# c. Supremum Distance (h=inf)
dist_supremum = squareform(pdist(subset_data, metric='chebyshev'))

# 3. Tampilkan Hasil Matriks Dissimilarity (Euclidean)
dist_df = pd.DataFrame(dist_euclidean,
                       columns=[f'Obj {i+1}' for i in range(5)],
                       index=[f'Obj {i+1}' for i in range(5)])

print("Data Matrix (5 baris pertama):")
print(subset_data)
print("\nDissimilarity Matrix (Euclidean):")
print(dist_df)
```

**penjelasan code**

1. Persiapan Data (Load Data) Kode dimulai dengan memuat dataset Iris menggunakan sklearn datasets Atribut Numerik: Dataset ini terdiri dari 4 dimensi (sepal length, sepal width, petal length, petal width)

    Data Matrix: Baris data diubah menjadi struktur matriks (Pandas DataFrame) agar bisa dihitung jarak antar barisnya (objek).

2.  Perhitungan jarak minkowski Distance, jarak untuk data numerik dihitung menggunakan rumus Minkowski Distance. Di dalam kode, kita menggunakan fungsi pdist (Pairwise Distance) dari pustaka scipy:Manhattan Distance ($h=1$): Menggunakan parameter metric='cityblock'. Ini menghitung jumlah selisih absolut antar atribut (L1 norm).Euclidean Distance ($h=2$): Menggunakan parameter metric='euclidean'. Ini adalah ukuran jarak garis lurus standar (L2 norm) Supremum Distance ($h \to \infty$): Menggunakan parameter metric='chebyshev'. Ini hanya mengambil selisih nilai maksimum di antara semua atribut yang ada dalam satu vektor.

3.  Pembentukan dissimilarity matrixHasil dari perhitungan jarak tersebut biasanya berupa array satu dimensi. Untuk mengubahnya menjadi tampilan tabel yang bisa dibaca melakukan langkah berikut:
squareform: Fungsi ini mengubah array jarak menjadi matriks Segitiga yang simetris.Sifat Matriks:nilai diagonal selalu 0 karena jarak objek ke dirinya sendiri adalah nol ($d(i,i)=0$).Matriks ini bersifat Simetri, artinya jarak dari objek 1 ke 2 sama dengan jarak dari objek 2 ke 1 ($d(i,j) = d(j,i)$).

4. Interpretasi hasil setelah kode dijalankan, Anda akan mendapatkan tabel (dataFrame) yang menunjukkan nilai numerik perbedaan antar bunga Iris:nilai rendah: menunjukkan dua objek sangat mirip (similarity tinggi).nilai tinggi: menunjukkan dua objek memiliki perbedaan besar (dissimilarity tinggi).
